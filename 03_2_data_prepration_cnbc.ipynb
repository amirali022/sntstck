##### Import statements

```python
import pandas as pd
from utils import merge_historical_with_sentiment
```

##### Load news data

```python
cnbc = pd.read_csv( "data/result/cnbc-data-with-sentiment.csv", sep=";")
```

```python
cnbc.head()
```

##### Group sentiment scores by date (averaging scores)

```python
dates = []

for i in range( len( cnbc)):
	dates.append( cnbc[ "published_date"][ i][ :10])

cnbc[ "Date"] = dates
```

```python
cnbc_groupby_date = cnbc[ [ "Date", "neg", "neu", "pos", "compound"]].groupby( "Date").mean().reset_index()
```

```python
cnbc_groupby_date.head()
```

##### Prepare S&P 500

```python
sp500_data = pd.read_csv( "data/input/sp500.csv")

sp500_data.head()
```

```python
sp500_with_sentiment = merge_historical_with_sentiment( sp500_data, cnbc_groupby_date)
```

```python
sp500_with_sentiment.head()
```

```python
sp500_with_sentiment.to_csv( "data/result/sp500-with-cnbc-sentiment.csv")
```
