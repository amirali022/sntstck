##### Import statements

```python
import pandas as pd
from utils import merge_historical_with_sentiment
```

##### Load news data

```python
forbes = pd.read_csv( "data/result/forbes-data-with-sentiment.csv", sep=";")
```

```python
forbes.head()
```

##### Group sentiment scores by date (averaging scores)

```python
dates = []

for i in range( len( forbes)):
	dates.append( forbes[ "date"][ i][ :10])

forbes[ "Date"] = dates
```

```python
forbes_groupby_date = forbes[ [ "Date", "neg", "neu", "pos", "compound"]].groupby( "Date").mean().reset_index()
```

```python
forbes_groupby_date.head()
```

##### Prepare S&P 500

```python
sp500_data = pd.read_csv( "data/input/sp500.csv")

sp500_data.head()
```

```python
sp500_with_sentiment = merge_historical_with_sentiment( sp500_data, forbes_groupby_date)
```

```python
sp500_with_sentiment.head()
```

```python
sp500_with_sentiment.to_csv( "data/result/sp500-with-forbes-sentiment.csv")
```
