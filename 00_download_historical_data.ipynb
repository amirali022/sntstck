```python
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
def plot_history( data, title):
	plt.figure( figsize=( 10, 6))
	plt.grid( True)
	plt.xlabel( "Dates")
	plt.ylabel( "Close Price")
	plt.plot( data[ "Close"])
	plt.title( title)
	plt.show()
```

#### S&P 500

```python
sp500 = yf.Ticker( "^GSPC")

sp500_data = sp500.history(
	period="1d",
	interval="1d",
	start="2020-08-01",
	end="2024-03-08"
)
```

```python
sp500_data.head( 3)
```

```python
sp500_data.tail( 3)
```

```python
plot_history( sp500_data, "S&P 500 Close Price")
```

```python
sp500_data.to_csv( "data/input/sp500.csv")
```
