##### Import statements

```python
import yfinance as yf
import matplotlib.pyplot as plt
```

##### Static Variables

```python
PERIOD = "1d"
START_DATE = "2018-01-01"
END_DATE = "2024-03-14"
```

##### Function for plotting close price time series

```python
def plot_history( data, title):
	plt.figure( figsize=( 10, 6))
	plt.grid( True)
	plt.xlabel( "Date")
	plt.ylabel( "Adjusted Close Price")
	plt.plot( data[ "Adj Close"])
	plt.title( title)
	plt.show()
```

##### Function for downloading historical data

```python
def historical( ticker_name, period = PERIOD, start = START_DATE, end = END_DATE):
	df = yf.download(
		ticker_name,
		period=period,
		start=start,
		end=end
	)

	return df
```

#### S&P 500

```python
sp500_data = historical( "^GSPC")
```

```python
sp500_data.head( 3)
```

```python
sp500_data.tail( 3)
```

```python
plot_history( sp500_data, "S&P 500")
```

```python
sp500_data.to_csv( "data/input/sp500.csv")
```
